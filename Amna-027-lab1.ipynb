# **LOOPS IN PYTHON**

# *For Loop*

```python
print("FOR LOOP:")
for i in range(1, 6):
    print(i)
```

# *While Loop*

```python
print("WHILE LOOP:")
i = 1
while i <= 5:
    print(i)
    i += 1
```

# *Nested Loop*

```python
print("NESTED LOOP:")
for i in range(1, 4):
    for j in range(1, 3):
        print(i, j)
```

# *Loop Using Break*

```python
print("LOOP WITH BREAK:")
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

# *Loop Using Continue*

```python
print("LOOP WITH CONTINUE:")
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

# *Conditional Loop*

```python
print("FOR LOOP WITH ELSE:")
for i in range(1, 4):
    print(i)
else:
    print("Loop finished successfully")
```

# *Infinite Loop (Controlled)*

```python
print("CONTROLLED INFINITE LOOP:")
i = 1
while True:
    print(i)
    if i == 3:
        break
    i += 1
```

# **FUNCTIONS IN PYTHON**

A function is a reusable block of code that performs a specific task and runs when it is called.

# *Simple Function (No Arguments, No Return)*

```python
def greet():
    print("Hello Python")

greet()
```

# *Built-in Function*

```python
numbers = [1, 2, 3, 4]
print(len(numbers))
```

# *User-Defined Function*

```python
def multiply(a, b):
    return a * b

print(multiply(4, 3))
```

# *Recursive Function*

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

# *Function with Arguments*

```python
def add(a, b):
    print("Sum =", a + b)

add(5, 3)
```

# *Function with Return Value*

```python
def square(num):
    return num * num

result = square(4)
print(result)
```

# *Function with Default Argument*

```python
def welcome(name="User"):
    print("Welcome", name)

welcome()
welcome("Amoou")
```

# *Function with Keyword Arguments*

```python
def student(name, age):
    print(name, age)

student(age=20, name="Ali")
```

# *Function with Variable-Length Arguments*

```python
def total(*numbers):
    print(sum(numbers))

total(1, 2, 3, 4)
```

# *Function with Variable-Length Keyword Arguments*

```python
def info(**data):
    print(data)

info(name="Ali", age=21)
```

# **DICTIONARIES IN PYTHON**

A dictionary is a collection of data stored as key–value pairs, where each key is unique.

```python
student = {"name": "Ali", "age": 20, "marks": 85}
print(student)
```

# *Access Values*

```python
print(student["name"])
```

# *List ➝ Dictionary (using zip)*

```python
keys = ["name", "age", "grade"]
values = ["Sara", 18, "A"]

data = dict(zip(keys, values))
print(data)
```

# *List of Tuples ➝ Dictionary*

```python
items = [("a", 1), ("b", 2), ("c", 3)]
d = dict(items)
print(d)
```

# *Dictionary ➝ List (Keys)*

```python
print(list(student.keys()))
```

# *Dictionary ➝ List (Values)*

```python
print(list(student.values()))
```

# *Dictionary ➝ List (Key–Value Tuples)*

```python
print(list(student.items()))
```

# *Tuple ➝ Dictionary*

```python
t = (("x", 10), ("y", 20))
d = dict(t)
print(d)
```

```python
print(student.get("marks"))
```

# **3. EXTENDED NUMPY OPERATIONS**

# *Matrix Multiplication*

Matrix multiplication is the operation of multiplying rows of the first matrix with columns of the second matrix using NumPy’s dot() or @ operator.

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])
```

```python
result = A @ B 
```

```python
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Result:\n", result)
```

# *Broadcasting*

Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array.

```python
result = A + B
```

```python
print("Array A:\n", A)
print("Array B:\n", B)
print("Result after Broadcasting:\n", result)
```
